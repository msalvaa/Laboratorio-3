## 8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

### La distancia entre dos coordenadas de la Tierra se puede calcular utilizando la fórmula del Haversine, que se muestra a continuación:

### R = radio de la Tierra
### Δlat = lat2− lat1
### Δlong = long2− long1
### a = sin²(Δlat/2) + cos(lat1) · cos(lat2) · sin²(Δlong/2)
### c = 2 · atan2(√a, √(1−a))
### d = R · c

```python
from math import sin, cos, sqrt, atan2, radians
R=6378
k=0
for i in df["Municipio"]:
    lat1=df[df["Municipio"]==i]["X"].values[0]
    lon1=df[df["Municipio"]==i]["Y"].values[0]
    lat2=df[df["Municipio"]=="Madrid"]["X"].values[0]
    lon2=df[df["Municipio"]=="Madrid"]["Y"].values[0]
    diflat=lat2-lat1
    diflon=lon2-lon1
    a=sin(diflat/2)**2+cos(lat1)*cos(lat2)*sin(diflon/2)**2
    c=2*atan2(sqrt(a),sqrt(1-a))
    d=R*c
    df.loc[k,"Distancia"]=d
    k=k+1
```

```python
max_d=df["Distancia"].max()
min_d=df[df["Distancia"]!=0]["Distancia"].min()
```

```python
print("El municipio más lejano es", df[df["Distancia"]==max_d]["Municipio"].values[0])
```

```python
print("El municipio más cercano es", df[df["Distancia"]==min_d]["Municipio"].values[0])
```
